```python
import os
import pandas
from datetime import datetime
import matplotlib.pyplot as plt
from pytz import utc
input_dir = os.path.join(os.getcwd(), "InputFiles")
data = pandas.read_csv(input_dir + r"\reviews.csv", parse_dates=["Timestamp"])
data.head()
data[data['Course Name'] == "The Python Mega Course: Build 10 Real World Applications"].count()
```

```python
data['Day'] = data["Timestamp"].dt.strftime("%m/%d/%Y")
average = data.groupby('Day').mean()
average.head()
plt.figure(figsize=(25,3))
plt.plot(average.index, average['Rating'])
```

```python
data['Week'] = data["Timestamp"].dt.strftime("%Y-%m")
average = data.groupby('Week').mean()
average.head()
```

```python
plt.figure(figsize=(25,3))
plt.plot(average.index, average['Rating'])
```

```python
data['Month'] = data["Timestamp"].dt.strftime("%Y-%m")
average = data.groupby(['Month', 'Course Name'])['Rating'].mean().unstack()
# print(average.head())

average.plot(figsize=(25,10))
```
